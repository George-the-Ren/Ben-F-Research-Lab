```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
final_dataset = pd.read_csv('/Users/Grant/Desktop/research/research_data.csv', index_col=0)
final_dataset.head()
```

```python
addresses = final_dataset['address']

geolocator = Nominatim(user_agent='research_geocode')

lat = []
long = []
a = []
for address in addresses: 
    try:
        location = geolocator.geocode({address})
        lat.append(location.latitude)
        long.append(location.longitude)
        a.append(location.address)
    except: 
        pass
```

```python
lat_long = {'latitude': lat, 'longitude': long, 'address': a}
df = pd.DataFrame(lat_long)
file_name = '/Users/Grant/Desktop/research/lat_long.csv'
df.to_csv(file_name, columns=['latitude', 'longitude', 'address'])
df.head(20)
```
