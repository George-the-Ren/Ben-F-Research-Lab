```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
from folium import IFrame
%matplotlib inline
```

```python
ls = '/Users/Grant/Desktop/research/lat_long_liquor.csv'
ss = '/Users/Grant/Desktop/research/lat_long_smoke_shops.csv'
s = '/Users/Grant/Desktop/research/lat_long_schools.csv'
m = '/Users/Grant/Desktop/research/lat_long_marijuana.csv'
liquor_stores = pd.read_csv(ls, index_col=0)
smoke_shops = pd.read_csv(ss, index_col=0)
schools = pd.read_csv(s, index_col=0)
marijuana = pd.read_csv(m, index_col=0)
```

```python
lat = 34
long = -118.35

# create map, plot points, and add a info icon
def create_map(data, text): 
    lat_long = data[['latitude', 'longitude']]
    address = data['address']
    list_df = lat_long.values.tolist()
    data_map = folium.Map(location=[lat, long], zoom_start=12)
    for point in range(0, len(list_df)):
        folium.CircleMarker(list_df[point], popup=address[point], 
                            radius=2, fill=True, color='blue').add_to(data_map)
    iframe = folium.IFrame(text, width=100, height=100)
    popup = folium.Popup(iframe, max_width=100)
    icon = folium.Marker(location=[lat,long], popup=popup, icon=folium.Icon(icon_color='green'))
    data_map.add_child(icon)
    return data_map
```

```python
ls_map = create_map(liquor_stores, 'Liquor Store Map')
ls_map.save('liquor_store_map.html')
ls_map
```

```python
ss_map = create_map(smoke_shops, 'Smoke Shops Map')
ss_map.save('smoke_shop_map.html')
ss_map
```

```python
s_map = create_map(schools, 'Schools Map')
s_map.save('schools_map.html')
s_map
```

```python
m_map = create_map(marijuana, 'Marijuana Dispensaries Map')
m_map.save('marijuana_map.html')
m_map
```
