```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
liquor_stores = pd.read_csv('/Users/Grant/Desktop/research/liquor_stores.csv', index_col=0)
smoke_shops = pd.read_csv('/Users/Grant/Desktop/research/smoke_shops.csv', index_col=0)
schools = pd.read_csv('/Users/Grant/Desktop/research/schools.csv', index_col=0)
marijuana = pd.read_csv('/Users/Grant/Desktop/research/marijuana.csv', index_col=0)
```

```python
def create_lat_long(data):
    addresses = data['address']
    geolocator = Nominatim(user_agent='lat long geocoder')
    lat = []
    long = []
    a = []
    for address in addresses: 
        try:
            location = geolocator.geocode({address})
            lat.append(location.latitude)
            long.append(location.longitude)
            a.append(location.address)
        except: 
            pass
    lat_long = {'latitude': lat, 'longitude': long, 'address': a}
    df = pd.DataFrame(lat_long)
    return df
```

```python
ls = create_lat_long(liquor_stores)
file_name = '/Users/Grant/Desktop/research/lat_long_liquor.csv'
ls.to_csv(file_name, columns=['latitude', 'longitude', 'address'])
ls.head(10)
```
