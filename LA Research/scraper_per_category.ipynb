```python
import json
import pandas as pd
import requests
import utils
```

```python
api_key = utils.api_key()
url = 'https://maps.googleapis.com/maps/api/place/textsearch/json?'

def dataframe_from_area_code(query):
    r = requests.get(url + 'query=' + query + '&key=' + api_key) 
    x = r.json() 
    y = x['results'] 
    df = pd.DataFrame(y)
    return df

global final_dataset

zip_code = ['90001', '90002', '90003', '90007', '90008', '90011', '90016',
            '90018', '90037', '90043', '90044', '90047', '90059', '90061',
            '90062', '90266', '90267', '90291', '90292', '90293', '90401', 
            '90402', '90403', '90404', '90405']
```

```python
smoke_shops = pd.DataFrame()

# dataset for smoke shops
for i in range(len(zip_code)):
    a = pd.DataFrame(dataframe_from_area_code('smoke shops near ' + zip_code[i]))
    smoke_shops = pd.concat([smoke_shops, a], axis=0, sort=True)
    
smoke_shops.drop(['id', 'icon', 'place_id', 'opening_hours', 
                    'plus_code', 'reference', 'photos', 'price_level', 
                    'rating', 'types', 'user_ratings_total'], axis=1, inplace=True)
smoke_shops.rename(columns = {'formatted_address': 'address', 'geometry': 'lat_long'}, inplace=True)
smoke_shops.drop_duplicates(subset='address', keep='first', inplace=True)
smoke_shops.reset_index(drop=True, inplace=True)
file_name = '/Users/Grant/Desktop/research/smoke_shops.csv'
smoke_shops.to_csv(file_name, columns = ['address', 'lat_long', 'name'])
print(smoke_shops.shape)
smoke_shops.head(10)
```
